## Let's code !

```python
# Set the active directory to the repos' root when locally testing (not necessary if MTM installed with pip or running on binder) 
%cd ..
```

```python
# 1st import the package and check its version
import MTM
print("MTM version: ", MTM.__version__)

from MTM import matchTemplates
from MTM.Detection import plotDetections

from skimage.data import coins

%matplotlib inline
import matplotlib.pyplot as plt
```

The main function in MTM is `matchTemplates`.  
Let's have a look at its docstring.

```python
matchTemplates?
```

As you can read from the doc, the function has 2 main image arguments :
- a list of template images to search
- an image in which to search for objects using the templates

Let's open an image of coins from scikit-image, we will showcase the multi-template matching for the detection of the coins' locations

```python
image = coins()
plt.imshow(image, cmap="gray")
```

## - crop a template  
We need some template image to search in the image.  
We can simply crop some coins from the image.  
__NB : Templates are always rectangular (they are image matrices), and thus the predicted locations are always rectangular too !__

```python
smallCoin = coins()[37:37+38, 80:80+41] 
plt.imshow(smallCoin, cmap="gray")
```

## - Calling matchTemplates with a single template
We will now use the `matchTemplates` that we have seen above.

```python
listTemplate = [smallCoin]

# Then call the function matchTemplates (here a single template)
listDetections = matchTemplates(image, 
                                listTemplate, 
                                score_threshold=0.5, 
                                maxOverlap=0)
```

Here the coins are nicely separated so we do not expect any overlap between the predicted coins locations ie `maxOverlap=0`.  
Let's have a look at `listDetections`...

```python
print("Found {} hits".format( len(listDetections) ) )
listDetections
```

The function `matchTemplates` returns a list of BoundingBoxes.  
BoundingBoxes are a sub-type of the Detection object, which itself extend the shapely Polygon.  
A Detection (and thus BoundingBoxes) contains :  
- the correlation score with the template used for this search  
- the coordinates (in pixels) of the bounding box in the image (x, y, width, height), with (x,y) the coordinates of the upper left corner  
- the positional index of the corresponding template in the original `listTemplate`

If we had provided a `listLabels` to `matchTemplates`, each hit would also have its corresponding label.  
The first hit in the list has a score of 1 and is the coin that is used as template.

## - View the detections
We can then represent the detections as bounding boxes overlaid on the image.  
The detections are color-coded according to the template that was matching for this detection.     

```python
plotDetections(image, listDetections)
```

## - Decreasing the score-threshold
This is pretty good already but we dont catch some of the small coins.  
Let's reduce the score threshold to be more permissive e.g. `score_threshold = 0.4` and display the score on the figure using the argument `showScore=True`

```python
listDetections = matchTemplates(image, listTemplate, score_threshold=0.4, maxOverlap=0)
plotDetections(image, listDetections, showScore=True)
```

## - Using multiple templates
Using a single template is quite good already for this particular example.    
What if now we would like to do better and somehow differentiate between small and large coins...  
The function matchTemplates allow to use several templates, and yields the best matching template for each detection in the image. 

Let's crop one of the big coin to use as a second template.

```python
largeCoin = coins()[14:14+59,302:302+65]
plt.figure(0)
plt.imshow(smallCoin, cmap="gray")
plt.figure(1)
plt.imshow(largeCoin, cmap="gray")
```

Again we pack the templates in a list before calling the function `matchTemplates` and displaying the result.

```python
listTemplates = [smallCoin, largeCoin]
listDetections = matchTemplates(image, listTemplates, score_threshold=0.4, maxOverlap=0)
plotDetections(image, listDetections)
```

Et voila !  
Using several templates can be used to increase the chance to catch your object if it appears with different perspectives, or to perform classification of the detections using different templates as "classes".  
However, the more templates the longer the computation time.
